```python
import os
import pandas as pd

# Define the folder path
folder_path = r'C:\Users\Pc\OneDrive - Aidwise Private Ltd\Desktop\11.27.2024'

# Initialize an empty list to store results
results = []

# Iterate through each file in the folder
for file in os.listdir(folder_path):
    if file.endswith(".xlsx") or file.endswith(".xls"):  # Check for Excel files
        file_path = os.path.join(folder_path, file)
        try:
            # Load the Excel file
            df = pd.read_excel(file_path)
            
            # Check if the MRN column exists
            if 'MRN' in df.columns:
                # Count distinct values in the MRN column
                unique_mrn_count = df['MRN'].nunique()
                results.append((file, unique_mrn_count))
            else:
                results.append((file, "No MRN column"))
        except Exception as e:
            results.append((file, f"Error: {e}"))

# Display the results
for file_name, count in results:
    print(f"{file_name}: {count} patients")
```

```python
import os
import pandas as pd


# Output DataFrame to store consolidated data
output_data = pd.DataFrame(columns=["File Name", "MRN", "Encounter DOS"])

# Iterate through each file in the folder
for file in os.listdir(folder_path):
    if file.endswith(".xlsx") or file.endswith(".xls"):  # Check for Excel files
        file_path = os.path.join(folder_path, file)
        try:
            # Load the Excel file
            df = pd.read_excel(file_path)
            
            # Check if both columns exist
            if {'MRN', 'Encounter DOS'}.issubset(df.columns):
                # Select required columns
                temp_df = df[['MRN', 'Encounter DOS']].copy()
                temp_df['File Name'] = file  # Add a column for the file name
                
                # Append to the output DataFrame
                output_data = pd.concat([output_data, temp_df], ignore_index=True)
        except Exception as e:
            print(f"Error processing file {file}: {e}")

# Save the consolidated data to an Excel file
output_file = os.path.join(folder_path, "consolidated_Encounter_data.xlsx")
output_data.to_excel(output_file, index=False)

print(f"Consolidated data saved to {output_file}")
```
